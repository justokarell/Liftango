### Steps

1. Reverse Geocode points
2. Add days of week (e.g. Mon, Tue, Wed, ...) into columns
3. Find weekday averages and distributions of stops; weekday_i = scenario_i (e.g. day 1, day 2, ... => scen 1, scen 2, ...)
4. Find distribution of hubs/stops for each day/scenario
5. Create Hub with the 'Locations', probabilities for 'day/scen 1', 'day/scen 2' , etc.
6. Create hub_scenario_probability by stacking the proabilties for each 'day/scen i' on one another
