```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
from datetime import datetime
from datetime import time
import seaborn as sns
```

```python
mtd = pd.read_csv('C://Users//12035//Downloads//2020-11 Matched Trip Details.csv')
mrs = pd.read_csv('C://Users//12035//Downloads//2020-11 Merged Ride Summary.csv')
ra = pd.read_csv('C://Users//12035//Downloads//2020-11 Ride Activity.csv')
rides = pd.read_csv('C://Users//12035//Downloads//rides_table.csv')
```

```python
location = 'Newcastle'

rides_mtd = mtd.loc[mtd['Ride Status'].isin(['accepted', 'merged'])]

tick_fontsize = 14
title_fontsize = 23
label_fontsize = 16
rides_mtd['Day'] = [a[8:10] for a in rides_mtd['Departure Time']]
days = sorted(list(set(rides_mtd['Day'])))
# days
rides_mtd.groupby('Day').count()['Car Make'].plot(figsize=(16,10), title='Matched Rides by Day')
plt.title(label='Matched Rides by Day', fontsize = title_fontsize, y=1.02)
plt.yticks(fontsize=tick_fontsize)
plt.xticks(ticks=np.arange(0,len(days)+1,5), labels=np.arange(0,len(days)+1,5), fontsize=tick_fontsize)
plt.xlabel('Day', fontsize=label_fontsize)
plt.grid(axis='y')


                                                  #.sort_values(['Departure Time']) #.plot(['Departure Time'])
```
