```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Read in our latitude longitude
''' Dataframe needs to have max lat max lon min lat min lon = the boundaries of the image file '''
df = pd.DataFrame({
'latitude': np.random.poisson(70,1000),
'longitude': np.random.poisson(70,1000),
'magnitude': np.random.normal(10,1000,1000)})


# Define a box-area based on the min and max of lat and long in our df 
BBox = ((df['longitude'].min(), df['longitude'].max(),df['latitude'].min(),df['latitude'].max()))

 
# filepath to your image
filepath = 'C:/Users/12035/Downloads/NYC_map.png'
# Load map image
ruh_m = plt.imread(filepath)


lat, lon = df['latitude'], df['longitude']
mag = df['magnitude']


# Define the figure, axes and plot
fig, ax = plt.subplots(figsize=(20,12))
sns.kdeplot(lat,lon, c=mag, shade=True, shade_lowest=False, alpha=0.35, cbar=False, ax=ax, cmap="viridis")
im = ax.scatter(lat,lon, c=mag, s=11.5, cmap='viridis')

fig.colorbar(im)

# Number of ticks on x and y axis
tick_num = 20

plot_title = "Plot Title"

plt.title(plot_title, fontsize=26, y=1.02)
plt.yticks(np.linspace(min(lon), max(lon), tick_num+1), fontsize=12, rotation=0)
plt.xticks(np.linspace(min(lat), max(lat), tick_num+1), fontsize=12, rotation=80)
plt.ylabel('Longitude', fontsize=14, labelpad=15)
plt.xlabel('Latitude', fontsize=14, labelpad=15)
ax.imshow(ruh_m, zorder=0, extent = BBox, aspect= 'equal')
plt.grid(True)
plt.show()
```
