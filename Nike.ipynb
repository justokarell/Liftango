```python

import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:

# nike_df['Passengers']
labels = list(Counter(nike_df['Passengers']).keys())
sizes = np.array(list(Counter(nike_df['Passengers']).values()))/sum(np.array(list(Counter(nike_df['Passengers']).values())))
explode = (0.2, 0.1, 0.3, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')
textprops = {"fontsize":25}
fig1, ax1 = plt.subplots(figsize=(16,15))
ax1.pie(sizes, explode=explode, 
        labels=labels, 
        autopct='%1.1f%%',
        shadow=True, 
        startangle=70,
       textprops = textprops)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
plt.figure(figsize=(5,10))
plt.boxplot(nike_df['Ride Time'].dropna())
plt.title("Ride Time Distribution", fontsize=22, y=1.02)
plt.ylabel('Time (min.)', fontsize=16)
plt.ylabel('')
plt.yticks(np.arange(0,nike_df['Ride Time'].max()+1))
plt.grid(axis='y')
plt.show()
```

```python

nike_df['Ride Time'] = abs(nike_df['Drop Off Time']-nike_df['Pickup Time'])
nike_df['Ride Time'] = [a.total_seconds()/60.0 for a in nike_df['Ride Time']]
nike_df['Ride Time'].hist(figsize=(16,10)
                          ,bins=np.arange(0,max(nike_df['Ride Time']),1.0) 
                          ,width=0.65 
                          ,align='mid'
                         )

plt.xticks(np.arange(0,max(nike_df['Ride Time']),1))
plt.yticks(np.arange(0,17,1))

plt.grid(axis='x')
plt.title("Ride Time Distribution", fontsize=22, y=1.02)
plt.show()

```

```python
D = dict(Counter(nike_df['Pick Up Location']))
B = dict(Counter(nike_df['Drop Off Location']))

B = dict(sorted(B.items(), key=lambda item: item[1]))
D = dict(sorted(D.items(), key=lambda item: item[1]))

plt.figure(figsize=(16,10))
plt.bar(range(len(D)), list(D.values()), align='center')
plt.title("Pick Up Location Distribution", fontsize=18)
plt.xticks(range(len(D)), list(D.keys()), rotation=90)

plt.figure(figsize=(16,10))
plt.bar(range(len(B)), list(B.values()), align='center')
plt.title("Drop Off Location Distribution", fontsize=18)
plt.xticks(range(len(B)), list(B.keys()), rotation=90)


plt.show()
```
