```python
times = []
for a in range(len(df.loc[(df['Codificação da Linha']==6)]['hora inicio viagem'])):
    ti = df.loc[(df['Codificação da Linha']==6)]['hora inicio viagem'].iloc[a]
    tf = df.loc[(df['Codificação da Linha']==6)]['hora final viagem'].iloc[a]
    times.append(np.array([tf.hour, tf.minute]) - np.array([ti.hour, ti.minute]))
ride_time = np.array([np.abs(a[0]-a[1]) for a in times]).mean()
```

```python
wait_times = []
for a in range(len(df.loc[(df['Codificação da Linha']==6)]['hora inicio viagem'])):
    ti = df.loc[(df['Codificação da Linha']==6)]['hora final viagem'].iloc[a-1]
    tf = df.loc[(df['Codificação da Linha']==6)]['hora inicio viagem'].iloc[a]
    wait_times.append(np.array([tf.hour, tf.minute]) - np.array([ti.hour, ti.minute]))
wait_avg = np.array([np.abs(a[0]-a[1]) for a in wait_times]).mean()
```

```python
def f1(X):
    return 0.29545*(X**0.4)*5 + 0.8

def f2(X):
    y = []
    for a in X:
        y.append(0.12545*(a**0.4)*4 + (np.random.choice([-1,0,0,2]))*np.random.random(6)[1])
    return np.array(y)
```

```python
FRT = np.append(3.5*np.ones(40),12*np.ones(9))
DRT = f2(np.arange(0,49))
diff = FRT - DRT

ave = np.mean(DRT)*np.ones(49)
fit = (np.mean(DRT)/49)*np.arange(49) + 1.0
plt.style.use('seaborn')


plt.figure(figsize=(14,10))
# plt.plot(diff, 'r--')
# plt.plot(FRT, 'g-', linewidth=2.6)
plt.plot(fit,'r-', linewidth=3.0)
plt.plot(ave,'g--')
plt.plot(DRT, 'b-')

plt.title("Average Distance Travelled (km) Per Shuttle: \n 6 Shuttles", fontsize=25)
plt.yticks(np.arange(0,11), fontsize=16)
plt.ylabel("Distance (km)", fontsize=20)
plt.xticks(np.arange(0,50,2), fontsize=16)
plt.xlabel("\nDemand \n (number of passengers)", fontsize=16)
plt.legend(['Fit','"Average"','DRT'],fontsize=18)
# plt.legend(['Difference','Fixed-Route','DRT'],fontsize=18)
```

```python
DRT = np.cumsum(np.array([0,6.5,4,1.5]))
FR = np.cumsum([0,6.5,6.5,6.5])
km_saved = FR-DRT
times = ['5:00pm', '5:45', '6:30pm', '7:15pm']

plt.style.use('seaborn')


plt.figure(figsize=(14,8))
plt.plot(times,DRT,c='b',linewidth=3.1)
plt.scatter(times,DRT,c='b',linewidth=3.1)
plt.plot(times, FR,c='g',linewidth=3.1)
plt.scatter(times, FR,c='g',linewidth=3.1)
plt.plot(times, km_saved, 'r--',linewidth=2.1)
plt.scatter(times, km_saved, c='r',linewidth=2.1)
plt.title("Afternoon Shift: \nCumulative Distance Travelled and Saved", fontsize=28, y=1.025)
plt.yticks(np.arange(0,40,2), fontsize=12)
plt.ylabel('Cumulative Distance Travelled (km) \n', fontsize=20)
plt.xticks(fontsize=14)
plt.grid(True)
plt.legend(['DRT','Fixed-Route','km Saved with DRT'], fontsize=18)
```
